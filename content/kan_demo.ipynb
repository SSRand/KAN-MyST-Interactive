```python
#| label: kan-fit

import math
import matplotlib.pyplot as plt
import torch
from kan import KAN, create_dataset

torch.set_default_dtype(torch.float64)
torch.manual_seed(0)

def target(x):
    return torch.exp(torch.sin(math.pi * x[:, [0]]) + x[:, [1]] ** 2)

dataset = create_dataset(target, n_var=2, train_num=300, test_num=300)
model = KAN(width=[2, 5, 1], grid=5, k=3, seed=1)

model.fit(dataset, opt="LBFGS", steps=20)

model.plot(beta=10, scale=0.5)
plt.gcf()
```

```python
#| label: kan-loss

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5.5, 3.2))
ax.plot(model.train_loss, label="train")
ax.plot(model.test_loss, label="test")
ax.set_yscale("log")
ax.set_xlabel("step")
ax.set_ylabel("loss")
ax.set_title("KAN training loss")
ax.legend()
fig.tight_layout()
fig
```
